```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('./dataset_ecommerce.csv')
df
```

```python
df["Método_Pagamento"].value_counts().plot(kind="pie", autopct="%1.1f%%")

plt.title("Distribuição dos Métodos de Pagamento")
plt.ylabel("")
plt.show()
```

```python
df.groupby("Categoria")["Receita_Total"].sum().plot(kind="bar")

plt.title("Receita por Categoria")
plt.xlabel("Categoria")
plt.ylabel("Receita")
plt.show()
```

```python
df.groupby("Produto")["Receita_Total"].sum().plot(kind="bar")

plt.title("Receita por Produto")
plt.xlabel("Produto")
plt.ylabel("Receita")
plt.show()
```

```python
media_produtos = df.groupby('Produto')['Quantidade_Vendida'].sum()
media_produtos.plot(kind='bar')
plt.title('Quantidade Vendida por Produto')
```

```python
df.groupby("Produto")["Nota_Avaliação"].mean().plot(kind="bar")

plt.title("Avaliação Média por Produto")
plt.xlabel("Produto")
plt.ylabel("Nota Média")
plt.show()
```

```python
plt.scatter(df["Quantidade_Vendida"], df["Receita_Total"])

plt.title("Quantidade Vendida vs Receita")
plt.xlabel("Quantidade Vendida")
plt.ylabel("Receita Total")
plt.show()
```

```python
plt.scatter(df["Preço_Unitário"], df["Quantidade_Vendida"])

plt.title("Preço vs Quantidade Vendida")
plt.xlabel("Preço Unitário")
plt.ylabel("Quantidade Vendida")
plt.show()
```

```python
df.boxplot(column="Preço_Unitário")
plt.title("Boxplot do Preço Unitário")
plt.show()
```
